**WEB SCRAPPING**

```python
import requests
import pandas as pd
import bs4
from bs4 import BeautifulSoup
```

```python
url="https://www.buscalibre.pe/libros/search?q=big+data"
```

```python
url=requests.get(url)
```

```python
cod_html=BeautifulSoup(url.content,"html.parser")
```

**Estructurar**

```python
cod_html
```

```python
nombres=cod_html.find_all("h3",{"class":"nombre"})
```

```python
nombre_list=[nombres[i].get_text() for i in range(0,len(nombres))]
```

```python
nombre_list
```

```python
dicc_libro ={"nombre del libro":nombre_list}
```

```python
dicc_libro
```

```python
df_libros =pd.DataFrame(dicc_libro)
```
